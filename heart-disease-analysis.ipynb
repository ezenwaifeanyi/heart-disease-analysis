## Matplotlib Example workflow

```python
# step 0.  import matplotlib and get it ready for plotting in jupyter
%matplotlib inline 
import matplotlib.pyplot as plt

# step 1.  Prepare data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

# step 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 10)) # (width, height)

# step 3. Plot data
ax.plot(x, y)

# step 4. Customize plot
ax.set(title="Simple Plot", xlabel = "x-axis", ylabel = "y-axis")

# step 5. save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Which one should we use? (pyplot vs matplotlib 00 method?)
When plotting something quickly, okay to use the pyploy method(recall method 1 and 2 above are pyplot)
When plotting something more advanced use the 00 method(method 3 above is  00method)

```python
heat_disease.head()
```

```python
# getting age over 50 from data set
over_50 = heat_disease[heat_disease["age"]> 50]
over_50
```

```python
# pyplot method
over_50.plot(kind = 'scatter', x = 'age', y= 'chol', c = 'target');
```

```python
# oo method (recommended and more advance)

fig, ax= plt.subplots(figsize=(10, 6))
over_50.plot(kind ='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim((45, 100));
```

```python
# 00 method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data

scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])

# Customize the plot

ax.set(title= "Heat Disease and Cholesterol Levels",
      xlabel= "Age",
      ylabel= "Cholesterol")

#Add a legend

ax.legend(*scatter.legend_elements(), title= "Target");

# Add a horizontal line

ax.axhline(over_50["chol"].mean(),
          linestyle = '--');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize = (10,10), 
                              sharex = True)

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])
#customize
ax0.set(title= "Heat Disease and Cholesterol Levels",
      ylabel= "Cholesterol")

# add legend
ax0.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line

ax0.axhline(over_50["chol"].mean(),
          linestyle = '--');



# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                    y = over_50["thalach"],
                    c = over_50["target"])
#customize ax1
ax1.set(title= "Heat Disease and Max Heart Rate",
      xlabel= "Age",
      ylabel= "Max Heat Rate")

# add legend ax1
ax1.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line

ax1.axhline(over_50["thalach"].mean(),
          linestyle = '--');

# Add a title to the figure
fig.suptitle("Heat Disease Analysis", fontsize=16, fontweight="bold");
```

## This plot shows some infomation about the heart disease

```python
# Customizing the x and y axis limitations

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize = (10,10), 
                              sharex = True)

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"],
                     cmap = "winter")
#customize ax0
ax0.set(title= "Heat Disease and Cholesterol Levels",
      ylabel= "Cholesterol")
# change the x axis limits
ax0.set_xlim([50,80])
# add legend
ax0.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line

ax0.axhline(over_50["chol"].mean(),
          linestyle = '--');



# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                    y = over_50["thalach"],
                    c = over_50["target"],
                    cmap = "winter")
#customize ax1
ax1.set(title= "Heart Disease and Max Heart Rate",
      xlabel= "Age",
      ylabel= "Max Heat Rate")
# Change ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# add legend ax1
ax1.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line

ax1.axhline(over_50["thalach"].mean(),
          linestyle = '--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
